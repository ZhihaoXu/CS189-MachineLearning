```python
import numpy as np
import matplotlib.pyplot as plt
from imageio import imread

%matplotlib inline 
```

# Part a)

```python
face = imread("./data/face.jpg")
plt.set_cmap('gray')
plt.imshow(face)
```

```python
def low_rank_approximation(image, r):
    u, s, v = np.linalg.svd(image,full_matrices=False)
    s[r:] = 0
    return np.dot(u * s, v)
```

```python
face_lr5 = low_rank_approximation(face, 5)
plt.subplot(121)
plt.imshow(face)
plt.title('Original image')
plt.subplot(122)
plt.imshow(face_lr5)
plt.title('low rank=%d' % 5)
plt.show()
```

```python
face_lr20 = low_rank_approximation(face, 20)
plt.set_cmap('gray')
plt.subplot(121)
plt.imshow(face)
plt.title('Original image')
plt.subplot(122)
plt.imshow(face_lr20)
plt.title('low rank=%d' % 20)
plt.show()
```

```python
face_lr100 = low_rank_approximation(face, 100)
plt.subplot(121)
plt.imshow(face)
plt.title('Original image')
plt.subplot(122)
plt.imshow(face_lr100)
plt.title('low rank=%d' % 100)
plt.show()
```

# Part c)

```python
def mse(img1, img2):
    return np.sum(np.square(img1-img2))
```

```python
allFaceMSE = []
for r in range(100):
    img_lr = low_rank_approximation(face, r+1)
    allFaceMSE.append(mse(img_lr,face))
plt.title("All Face Low-Rank Approximation Mse")
plt.plot(allFaceMSE)
```

```python
allSkyMSE = []
for r in range(100):
    img_lr = low_rank_approximation(sky, r+1)
    allSkyMSE.append(mse(img_lr,sky))
plt.title("All Sky Low-Rank Approximation Mse")
plt.plot(allSkyMSE)
```
